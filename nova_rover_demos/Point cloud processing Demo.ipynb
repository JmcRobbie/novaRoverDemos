This notebook will briefly demonstrate the point cloud processing/ generating class.

```python
from pointCloudProcessing.ptcloudgen import point_cloud_gen
```

```python
size = [(0,1),(0,2),(0,5)] ## Size of data read in
```

```python
cld = point_cloud_gen(size,4) ## Generating class instance with default number of points and 4 features
cld.gaussianCloud()
cld.GetOccupancyGrid(0.05,0.05) ## 5cm grid resolution
cld.plotOccGrid()
cld.plotCloud()
```
